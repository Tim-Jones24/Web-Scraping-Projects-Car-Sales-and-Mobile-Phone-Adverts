## Mobile Phone Adverts Scraping from an E-Commerce Website

### Import Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### Extract Relevant Content from the Website

```python
url= 'https://jiji.ng/mobile-phones'
response= requests.get(url)

# Number of pages to extract
no_of_pages= 200

phone_data=[]
for page in range (1,no_of_pages+1):
    url= f'(url)?page=(page)'
    if response.status_code == 200:
        sp= BeautifulSoup(response.content, 'html.parser')
        #print(sp)

        phones_data= sp.find_all('div', class_ = 'b-list-advert-base__data')


        for phone in phones_data:
            #Extract mobile names
            Name= phone.find('div', class_ = 'b-list-advert-base__data__title')
            phone_name= Name.text.strip() if Name else 'N/A'

            #Extract Type names
            Type= phone.find('div', class_ = 'b-list-advert-base__attrs')
            phone_type= Type.text.strip() if Type else 'N/A'

            #Extract Description
            Desc= phone.find('div', class_ = 'b-list-advert-base__description')
            phone_desc= Desc.text.strip() if Desc else 'N/A'

            #Extract price 
            Price= phone.find('div', class_ = 'b-list-advert-base__data__price')
            phone_price= Price.text.strip() if Price else 'N/A'

            #Append to a list
            phone_data.append(
                {'Mobile': phone_name,
                 'Type': phone_type,
                 'Description': phone_desc,
                 'Price': phone_price        
            })
    else: None
else:
    print('',response.status_code)

    
#Converting into a DataFrame
df= pd.DataFrame(phone_data)
df   

    

        
        
```

### Save to CSV File

```python
df.to_csv('phone_data.csv',index_label=False)
```
